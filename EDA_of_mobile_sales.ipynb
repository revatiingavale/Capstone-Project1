```python
#plot the histogram of price
sns.histplot(data['Amount'])
plt.show()
```

```python
#plot scatter plot
sns.scatterplot(x='Quantity', y='Amount', data=data)
plt.title('Quantity vs. Amount')
plt.xlabel('Quantity')
plt.ylabel('Amount')
plt.show()
```

```python
#plot pie chart
paymentcount=data["PAYMENT TYPE"].value_counts()
plt.pie(paymentcount,labels=paymentcount.index,autopct='%1.1f%%',startangle=140)
plt.show()
```

```python
sns.barplot(x="PAYMENT TYPE",y="Amount",data=data)
plt.show()
```
